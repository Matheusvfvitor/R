```
report = read.csv("2019_nCoV_data.csv")
head(report)
```

```
report

str(report)
print("-------")
summary(report)
```

```
library(tidyr)
library(dplyr)
```

```
rp_smmy = report %>% filter(Confirmed > 0)

ggplot(rp_smmy,aes(Country, fill = Deaths))+
geom_bar()+
theme_classic()
```

```
rp_smmy = report %>% filter(Confirmed > 0 , Country != "Mainland China")

ggplot(rp_smmy,aes(Country, fill = Deaths))+
geom_bar()+
theme_classic()+
theme(axis.text = element_text(angle=90, color ="blue", face = 3))
```

```
rp_smmy = report %>% group_by(Date) %>% summarise(Total = n()) 

rp_smmy
```
